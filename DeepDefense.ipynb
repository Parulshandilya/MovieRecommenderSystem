```python
import csv
import pandas as pd
import os
import xml.etree.cElementTree as et
```

```python
data_xml = {
    "appName": [],
    "totalSourceBytes": [],
    "totalDestinationBytes": [],
    "totalDestinationPackets": [],
    "totalSourcePackets": [],
    "sourcePayloadAsBase64": [],
    "destinationPayloadAsBase64": [],
    "destinationPayloadAsUTF": [],
    "direction": [],
    "sourceTCPFlagsDescription": [],
    "destinationTCPFlagsDescription": [],
    "source": [],
    "protocolName": [],
    "sourcePort": [],
    "destination": [],
    "destinationPort": [],
    "startDateTime": [],
    "stopDateTime": [],
    "Tag": []
    
}
```

```python
parsed_xml = et.parse("labeled_flows_xml/TestbedMonJun14Flows.xml")
root = parsed_xml.getroot()
# for child in root:
for child in root:
    for child_temp in child:
        if (child_temp.text is not None):
            data_xml[child_temp.tag].append(child_temp.text)
        else:
            data_xml[child_temp.tag].append("")

df = pd.DataFrame(data=data_xml)    
print(df.head())
# print(df)
df.info()
print("------------------------------------")
df['Tag'].value_counts()
f = open('CSV_FILE.csv','w')
df.to_csv(f, header=True)
```

```python
for i in range(1,4):
    parsed_xml = et.parse("labeled_flows_xml/TestbedTueJun15-"+str(i)+"Flows.xml")
    root = parsed_xml.getroot()
    # for child in root:
    for child in root:
        for child_temp in child:
            if (child_temp.tag == "sensorInterfaceId" or child_temp.tag == "sourcePayloadAsUTF" ):
                continue
            elif (child_temp.text is not None):
                data_xml[child_temp.tag].append(child_temp.text)
            else:
                data_xml[child_temp.tag].append("")

    df = pd.DataFrame(data=data_xml)   
    df.info()
    print("------------------------------------")
    df['Tag'].value_counts()
    f = open('CSV_FILE.csv','a+')
    df.to_csv(f, header=False)
    f.close()
```
